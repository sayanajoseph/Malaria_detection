**TRAIN TEST SPLIT**

```python
x=df.iloc[:,:-1].values
x
```

```python
y=df.iloc[:,-1].values
y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
scaler.fit(x_train)
x_train=scaler.transform(x_train)
x_test=scaler.transform(x_test)
x_train
```

**HANDLING CLASS IMBALANCE**

```python
from imblearn.over_sampling import SMOTE
smote=SMOTE(random_state=42)
x_train,y_train=smote.fit_resample(x_train,y_train)
```

**BUILDING MODELS **

```python
from sklearn.neighbors import KNeighborsClassifier
knn=KNeighborsClassifier(n_neighbors=7)
from sklearn.svm import SVC
svm=SVC()
from sklearn.ensemble import RandomForestClassifier
rfc=RandomForestClassifier()
lst=[knn,svm,rfc]
```

**HYPERPARAMETER TUNING**

```python
from sklearn.model_selection import GridSearchCV
param_grid={'n_estimators':[100,200],'max_depth':[None,10,20]}
grid=GridSearchCV(rfc,param_grid,cv=5)
grid.fit(x_train,y_train)
best_rf=grid.best_estimator_
best_rf
```

**CROSS-VALIDATION ACCURACY**

```python
from sklearn.model_selection import cross_val_score
scores=cross_val_score(best_rf,x_train,y_train,cv=5)
scores.mean(),scores.std()
print(f"\n Accuracy: {scores.mean()},{scores.std()})")
```

**MODEL EVALUATION**

```python
y_pred=best_rf.predict(x_test)
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score,classification_report
```

```python
for i in lst:
  i.fit(x_train,y_train)
  y_pred=i.predict(x_test)
  print('model is',i)
  print(y_pred)
  print('confusion matrix')
  print(confusion_matrix(y_test,y_pred))
  print('accuracy score is')
  print(accuracy_score(y_test,y_pred))
  print('classification_report')
  print(classification_report(y_test,y_pred))
```

**ROC_AUC**

```python
from sklearn.metrics import roc_curve,roc_auc_score
```

```python
roc_auc_score=(y_test,best_rf.predict_proba(x_test)[:,1])
```

```python
fpr,tpr,thresholds=roc_curve(y_test,best_rf.predict_proba(x_test)[:,1])

plt.plot(fpr,tpr)
plt.xlabel("False positive rate")
plt.ylabel("True positive rate")
plt.title("ROC curve")
plt.show()
```
